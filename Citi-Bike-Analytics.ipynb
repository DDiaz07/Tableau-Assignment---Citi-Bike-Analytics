```python
import pandas as pd
```

Summery: I created a pattern with dataframes to make it easier to understand. I have data from 2013,2014,2015,2016,and 2017

# 2013 Data Set for Trip History

```python
df_one_2013 = pd.read_csv("2013_citibike.csv")
df_one_2013.head()
```

```python
df_two_2013 = pd.read_csv("2013_citibike_2.csv")
df_two_2013.head()
```

```python
df_three_2013 = pd.read_csv("2013_citibike_3.csv")
df_three_2013.head()
```

```python
frames = [df_one_2013, df_two_2013, df_three_2013]
```

```python
result_2013 = pd.concat(frames)
print (result_2013)
```

```python
#save on to csv
result_2013.to_csv('2013_citibike_results.csv')
```
